Question 1. Simulate 30 rolls with =RANDBETWEEN(1,6). What is the probability of rolling a 3 exactly 5 times? (Hint: Use BINOM.DIST)

- Answer 1

  To find the probability of rolling a 3 exactly 5 times out of 30 rolls:

  - Trials (n): 30
  - Successes (x): 5
  - Probability of success (p): 1/6 (since a die has 6 sides)
  - Excel Formula: =BINOM.DIST(5, 30, 1/6, FALSE)
  - Result: Approximately 0.1921 (or 19.21%).

Question 2. Generate 100 values in Excel using the continuous uniform distribution RAND() and plot a histogram. Describe the shape of the distribution.

- Answer 2


Process: Using =RAND() generates numbers between 0 and 1.

Shape: The histogram of 100 values should look roughly rectangular. Unlike a Bell Curve, every interval has an equal probability of occurring. As the sample size increases, the top of the histogram becomes flatter and more "uniform."

Question 4. What is the concept of standardization (z-score), and why is it important in data analysis? Explain the formula and how standardization transforms a dataset.

- Answer 4

Concept: Standardization shifts and scales data so that different datasets can be compared on the same "playing field."

Importance: It allows analysts to identify outliers and compare scores from different distributions (e.g., comparing an SAT score to an ACT score).

Question 5. What is Kurtosis and their type?

- Answer 5

Kurtosis measures the "tailedness" of a distribution (how much data is in the peaks vs. the tails).

Mesokurtic: Normal distribution (excess kurtosis = 0).

Leptokurtic: Sharp peak, fat tails (more outliers).

Platykurtic: Flat peak, thin tails (fewer outliers).


Question 6. Explain why the uniform distribution is a good model for the outcome of rolling a fair die.

- Answer 6.

A uniform distribution is a good model because a "fair" die implies that every outcome (1, 2, 3, 4, 5, 6) has an equal probability (1/6) of occurring. There is no bias toward any specific number, which is the definition of a discrete uniform distribution.


Quesion 7. Use Excel to compute the probability of getting at least 8 successes in 15 trials with success probability 0.5

- Answer 7

In Excel, BINOM.DIST with TRUE calculates the cumulative probability (from 0 up to x). To find "at least 8" (8 or more), you subtract the probability of 7 or fewer from 1.

Trials: 15 | Success Prob: 0.5Excel

Formula: =1 - BINOM.DIST(7, 15, 0.5, TRUE)

Result: Approximately 0.50 (or 50%).


Question 8. How does log transformation help in stabilizing variance and making data more normally distributed?

- Answer 8


Log transformation is used on right-skewed data (data with a long tail on the right, like income).

Stabilizing Variance: It compresses high values more than low values, which "reins in" outliers and spreads out the smaller values.

Normality: By pulling in the long right tail, it makes the distribution more symmetrical and bell-shaped, satisfying the assumptions of many statistical tests (like t-tests or linear regression).
